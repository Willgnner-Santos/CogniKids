1. Vis√£o Geral do Sistema
O CogniKids √© um prot√≥tipo funcional de um sistema de software educacional, desenvolvido para simular uma plataforma integrada de apoio √† inclus√£o de alunos neurodivergentes. O sistema foi concebido com uma arquitetura modular, centrada em tr√™s perfis de usu√°rio: Pais, Professores e Alunos. O objetivo principal do prot√≥tipo √© validar os fluxos de informa√ß√£o e as intera√ß√µes l√≥gicas entre esses tr√™s atores, demonstrando a viabilidade de um ecossistema digital colaborativo.

A aplica√ß√£o opera inteiramente em um ambiente de console, utilizando estruturas de dados em mem√≥ria para simular um banco de dados, o que significa que os dados n√£o s√£o persistentes entre as execu√ß√µes.

2. Arquitetura e Estruturas de Dados
O n√∫cleo do sistema √© sustentado por um conjunto de estruturas de dados globais que funcionam como um banco de dados em mem√≥ria. A escolha por esta abordagem visa a simplicidade e a agilidade na prototipagem.

Listas de Usu√°rios (vetores):

usuarios_pais = []: Armazena dicion√°rios, onde cada dicion√°rio representa o perfil de um pai/m√£e.
usuarios_professores = []: Armazena dicion√°rios com os perfis dos professores.
usuarios_alunos = []: Armazena dicion√°rios com os perfis dos alunos.
Estruturas de Dados de Intera√ß√£o:

perguntas_alunos = []: Uma lista de dicion√°rios que armazena cada pergunta feita por um aluno e a subsequente resposta do professor, criando um registro de comunica√ß√£o.
mensagens_pais_professores = []: Modela um sistema de mensagens diretas entre pais e professores.
forum_geral = []: Simula um f√≥rum de discuss√£o compartilhado entre pais e professores.
agenda_global = []: Armazena eventos, provas e atividades criadas pelo professor, vis√≠veis para os perfis relevantes.
Estruturas de Dados de Avalia√ß√£o:

pesos_por_turma = {}: Um dicion√°rio que armazena os pesos (prova, atividade, participa√ß√£o) definidos pelo professor para cada turma, essencial para o c√°lculo da m√©dia ponderada.
3. Descri√ß√£o dos M√≥dulos e Fun√ß√µes
O c√≥digo √© segmentado em fun√ß√µes que representam as principais a√ß√µes e telas do sistema.

3.1 Fun√ß√µes de Gerenciamento de Sess√£o

main(): √â a fun√ß√£o de entrada do programa. Cont√©m o loop principal que apresenta a tela de boas-vindas, gerencia a sele√ß√£o de perfil (Pai, Aluno, Professor) e direciona o usu√°rio para o fluxo de cadastro ou de login.
realizar_login(): Solicita usu√°rio e senha, itera sobre todas as listas de usu√°rios e valida as credenciais. Retorna o dicion√°rio do usu√°rio logado em caso de sucesso ou None em caso de falha.
cadastrar_pai(), cadastrar_aluno(), cadastrar_professor(): Fun√ß√µes espec√≠ficas que coletam os dados necess√°rios para cada perfil e os adicionam √† lista global correspondente. A fun√ß√£o cadastrar_aluno inclui a l√≥gica para a sele√ß√£o de avatar via menu.
3.2 M√≥dulo do Professor (tela_professor)

Este m√≥dulo centraliza as ferramentas de gest√£o acad√™mica e comunica√ß√£o do professor.

Acesso a Dados de Apoio: Consulta os perfis dos pais (usuarios_pais) para extrair e exibir informa√ß√µes qualitativas sobre o aluno (neurodiverg√™ncia, Kit de Apoio).
Defini√ß√£o de Pesos: Permite ao professor atribuir pesos num√©ricos para "Provas", "Atividades" e "Participa√ß√£o" para cada uma de suas turmas. Esses dados s√£o armazenados no dicion√°rio pesos_por_turma.
Lan√ßamento de Notas: Implementa um fluxo sequencial e otimizado para que o professor, ap√≥s selecionar um aluno, possa inserir notas de prova, atividade e participa√ß√£o, al√©m de faltas, em uma √∫nica tela. Os valores inseridos s√£o adicionados √†s listas correspondentes dentro do dicion√°rio de desempenho do aluno.
Visualiza√ß√£o de Desempenho: Chama a fun√ß√£o exibir_grafico_desempenho() para o aluno selecionado.
Painel de Sentimentos: Acessa o perfil do aluno e exibe o hist√≥rico de registros emocionais (emoji e texto) inseridos por ele.
Responder Perguntas: Filtra a lista global perguntas_alunos para exibir apenas as perguntas que ainda n√£o possuem uma chave "resposta". Ap√≥s o professor inserir a resposta, a fun√ß√£o atualiza o dicion√°rio original da pergunta, adicionando a resposta.
3.3 M√≥dulo do Aluno (tela_aluno)

Interface focada na experi√™ncia do estudante, com elementos de gamifica√ß√£o e autoexpress√£o.

Painel de Sentimentos: Apresenta um menu de emojis para que o aluno escolha o que melhor representa seu estado emocional. Em seguida, solicita um texto descritivo. Ambos os dados s√£o salvos, com data, no perfil do aluno, ficando dispon√≠veis para consulta pelos pais e professor.
Desafios do Sol: Exibe uma lista de 10 desafios pr√©-definidos. Ao selecionar um desafio, o contador desafios_sol no perfil do aluno √© incrementado. O progresso √© exibido visualmente atrav√©s de uma "barra de progresso" com emojis de sol (‚òÄÔ∏è).
Perguntas e Respostas: O aluno pode "Fazer uma pergunta", que √© adicionada √† lista perguntas_alunos. Na op√ß√£o "Ver minhas Perguntas e Respostas", o sistema filtra essa lista global, exibindo apenas as perguntas feitas pelo aluno logado e a resposta do professor, caso exista.
3.4 M√≥dulo dos Pais (tela_pais)

Funciona como um painel de controle e acompanhamento, integrando as informa√ß√µes dos outros m√≥dulos.

Kit de Apoio e Neurodiverg√™ncia: Permite aos pais inserir e atualizar informa√ß√µes qualitativas essenciais sobre o filho, que ficam dispon√≠veis para o professor.
Visualiza√ß√£o de Dados: Todas as fun√ß√µes de visualiza√ß√£o (Agenda, Painel de Sentimentos, Desempenho, Perguntas/Respostas) funcionam como "leitores" das estruturas de dados globais. Elas filtram as informa√ß√µes relevantes para o filho do pai logado e as exibem de forma formatada.
3.5 Fun√ß√£o de Desempenho (exibir_grafico_desempenho)

Esta √© uma fun√ß√£o-chave, chamada tanto pelo professor quanto pelos pais e aluno.

Recebe um dicion√°rio de aluno como par√¢metro.
Busca no dicion√°rio pesos_por_turma os pesos correspondentes √† turma do aluno.
Calcula a m√©dia aritm√©tica simples de cada crit√©rio (provas, atividades, participa√ß√£o) a partir das listas de notas do aluno.
Aplica a f√≥rmula da m√©dia ponderada: ((media_provas * peso_p) + ...) / soma_pesos.
Imprime uma matriz formatada que detalha o c√°lculo, mostrando a m√©dia de cada crit√©rio, o peso aplicado e a contribui√ß√£o daquele crit√©rio para a nota final, al√©m da M√©dia Ponderada Final.

## Funcionalidades Principais

```python
# -*- coding: utf-8 -*-
import time
import math
import random
import textwrap

# ==============================================================================
# ESTRUTURA DE DADOS (SIMULANDO UM BANCO DE DADOS EM MEM√ìRIA)
# ==============================================================================

usuarios_pais = []
usuarios_alunos = []
usuarios_professores = []
agenda_global = []
forum_geral = []
mensagens_pais_professores = []
perguntas_alunos = []
galeria_infantil = []
pesos_por_turma = {}

# ==============================================================================
# FUN√á√ïES AUXILIARES
# ==============================================================================

def limpar_tela():
    print("\n" * 50)

def aguardar_enter():
    input("\nPressione Enter para continuar...")

def formatar_texto_longo(texto, largura=65, recuo_inicial='  ', recuo_subsequente='  '):
    if not texto:
        return recuo_inicial + "N/A"
    linhas_quebradas = textwrap.wrap(texto, width=largura)
    texto_formatado = recuo_inicial + ("\n" + recuo_subsequente).join(linhas_quebradas)
    return texto_formatado

def encontrar_aluno_por_nome(nome_aluno):
    for aluno in usuarios_alunos:
        if aluno['nome_completo'].lower() == nome_aluno.lower():
            return aluno
    return None

def encontrar_pai_por_filho(nome_filho):
    for pai in usuarios_pais:
        if nome_filho in pai['filhos_nomes']:
            return pai
    return None

# ==============================================================================
# FUN√á√ïES DE CADASTRO
# ==============================================================================

def cadastrar_pai():
    limpar_tela()
    print("--- Cadastro de Pai/M√£e ---")
    nome_completo = input("Nome completo: ")
    filhos_str = input("Nomes dos filhos (separados por v√≠rgula): ")
    filhos_nomes = [nome.strip() for nome in filhos_str.split(',')]
    neurodivergencias = input("Neurodiverg√™ncias de cada um (se houver, ou pressione Enter): ")
    instituicao = input("Institui√ß√£o de ensino dos filhos: ")
    username = input("Crie um nome de usu√°rio: ")
    senha = input("Crie uma senha: ")
    usuarios_pais.append({
        'tipo': 'pai', 'nome_completo': nome_completo, 'filhos_nomes': filhos_nomes,
        'neurodivergencias': neurodivergencias, 'instituicao': instituicao,
        'username': username, 'senha': senha, 'dados_filhos': {}
    })
    print("\nCadastro de Pai/M√£e realizado com sucesso!")
    aguardar_enter()

def cadastrar_aluno():
    limpar_tela()
    print("--- Cadastro de Aluno(a) ---")
    nome_completo = input("Seu nome completo: ")
    nome_pais = input("Nome do seu pai ou m√£e: ")
    nome_avatar = input("Escolha um nome para seu amigo Avatar: ")

    print("\nEscolha seu amigo Avatar na lista abaixo:")
    opcoes_avatar = {
        '1': 'ü¶Ñ Unic√≥rnio M√°gico', '2': 'üêâ Drag√£o Amigo', '3': 'ü¶ñ Dinossauro Rex',
        '4': 'ü¶Å Le√£o Corajoso', '5': 'ü¶ä Raposa Esperta', '6': 'üê∂ Cachorrinho Leal',
        '7': 'üêº Panda Fofinho', '8': 'üê¨ Golfinho Brincalh√£o', '9': 'ü¶ã Borboleta Colorida',
        '10': 'ü§ñ Rob√¥ Curioso'
    }
    for numero, descricao in opcoes_avatar.items():
        print(f"{numero}. {descricao}")
    escolha_avatar = input("\nDigite o n√∫mero do seu avatar escolhido: ")
    imagem_avatar_desc = opcoes_avatar.get(escolha_avatar, "‚≠ê Estrela Amiga")

    username = input("Crie um nome de usu√°rio: ")
    senha = input("Crie uma senha: ")
    usuarios_alunos.append({
        'tipo': 'aluno', 'nome_completo': nome_completo, 'nome_pais': nome_pais,
        'nome_avatar': nome_avatar, 'imagem_avatar': imagem_avatar_desc,
        'username': username, 'senha': senha, 'painel_sentimentos': [],
        'desempenho': {'provas': [], 'atividades': [], 'participacao': [], 'faltas': 0},
        'desafios_sol': 0,
        'turma': input("Qual a sua turma? (Ex: 3A): ")
    })
    print(f"\nCadastro de Aluno(a) realizado com sucesso! Ol√°, {nome_avatar}!")
    print(f"Seu novo amigo √© o {imagem_avatar_desc}!")
    aguardar_enter()

def cadastrar_professor():
    limpar_tela()
    print("\n--- Cadastro de Professor(a) ---")
    nome_completo = input("Nome completo: ")
    instituicao = input("Institui√ß√£o de ensino onde trabalha: ")
    turmas_str = input("Turmas que leciona (separadas por v√≠rgula): ")
    turmas = [turma.strip() for turma in turmas_str.split(',')]
    username = input("Crie um nome de usu√°rio: ")
    senha = input("Crie uma senha: ")
    usuarios_professores.append({
        'tipo': 'professor', 'nome_completo': nome_completo, 'instituicao': instituicao,
        'turmas': turmas, 'username': username, 'senha': senha
    })
    print("\nCadastro de Professor(a) realizado com sucesso!")
    aguardar_enter()

# ==============================================================================
# FUN√á√ÉO DE LOGIN
# ==============================================================================

def realizar_login():
    limpar_tela()
    print("========= P√ÅGINA DE LOGIN =========")
    username = input("Usu√°rio: ")
    senha = input("Senha: ")

    todos_usuarios = usuarios_pais + usuarios_alunos + usuarios_professores
    for usuario in todos_usuarios:
        if usuario['username'] == username and usuario['senha'] == senha:
            print(f"\nLogin bem-sucedido! Bem-vindo(a), {usuario['nome_completo']}!")
            aguardar_enter()
            return usuario

    print("\nUsu√°rio ou senha inv√°lidos!")
    aguardar_enter()
    return None

# ==============================================================================
# FUN√á√ÉO DE GR√ÅFICO DE DESEMPENHO
# ==============================================================================

def exibir_grafico_desempenho(aluno):
    limpar_tela()
    print(f"--- Desempenho do Aluno: {aluno['nome_completo']} (Turma: {aluno['turma']}) ---")

    pesos = pesos_por_turma.get(aluno['turma'], {'provas': 0, 'atividades': 0, 'participacao': 0})
    soma_pesos = sum(pesos.values())

    if soma_pesos == 0:
        print("\nOs pesos para esta turma ainda n√£o foram definidos pelo professor.")
        aguardar_enter()
        return

    desempenho = aluno['desempenho']

    media_provas = sum(desempenho['provas']) / len(desempenho['provas']) if desempenho['provas'] else 0.0
    media_atividades = sum(desempenho['atividades']) / len(desempenho['atividades']) if desempenho['atividades'] else 0.0
    media_participacao = sum(desempenho['participacao']) / len(desempenho['participacao']) if desempenho['participacao'] else 0.0

    media_ponderada = (
        (media_provas * pesos['provas']) +
        (media_atividades * pesos['atividades']) +
        (media_participacao * pesos['participacao'])
    ) / soma_pesos

    matriz_desempenho = [
        ["Crit√©rio", "M√©dia (0-10)", "Peso Definido", "Contribui√ß√£o para a M√©dia"],
        ["Provas", f"{media_provas:.2f}", f"{pesos['provas']}", f"{(media_provas * pesos['provas'] / soma_pesos):.2f}"],
        ["Atividades", f"{media_atividades:.2f}", f"{pesos['atividades']}", f"{(media_atividades * pesos['atividades'] / soma_pesos):.2f}"],
        ["Participa√ß√£o", f"{media_participacao:.2f}", f"{pesos['participacao']}", f"{(media_participacao * pesos['participacao'] / soma_pesos):.2f}"]
    ]

    print("\nDEMONSTRATIVO DE C√ÅLCULO DA M√âDIA:\n")
    for linha in matriz_desempenho:
        print(f"{linha[0]:<15} | {linha[1]:<15} | {linha[2]:<15} | {linha[3]:<28}")

    print("\n" + "="*80)
    print(f"  M√âDIA PONDERADA FINAL: {media_ponderada:.2f}")
    print("="*80)
    print(f"\nInforma√ß√µes Adicionais:\n - Faltas Totais: {desempenho['faltas']}")
    aguardar_enter()

# ==============================================================================
# TELAS DOS USU√ÅRIOS (L√ìGICA COMPLETA E CORRIGIDA)
# ==============================================================================

def tela_pais(pai):
    while True:
        limpar_tela()
        print(f"========= PAINEL DOS PAIS - Ol√°, {pai['nome_completo']} =========")
        print("\n--- Gest√£o do Aluno ---")
        print("1. Inserir/Ver Kit de Apoio do Filho")
        print("2. Informar/Atualizar Neurodiverg√™ncia")
        print("3. Visualizar Agenda")
        print("4. Ver Painel de Sentimentos do Filho")
        print("5. Ver Desempenho Detalhado do Filho")
        print("6. Ver Perguntas e Respostas do Filho")
        print("\n--- Comunica√ß√£o ---")
        print("7. Visualizar F√≥rum")
        print("8. Postar no F√≥rum")
        print("9. Ler Mensagens")
        print("10. Enviar Mensagem para Professor")
        print("\n0. Logout (Sair)")
        opcao = input("Escolha uma op√ß√£o: ")

        if opcao == '1':
            nome_filho = input("Digite o nome do filho para o Kit de Apoio: ")
            if nome_filho in pai['filhos_nomes']:
                dados = pai['dados_filhos'].get(nome_filho, {})
                print(f"\n--- Kit de Apoio para {nome_filho} ---")
                print(f"1. Gatilhos de Crises: {dados.get('gatilhos', 'N√£o informado.')}")
                print(f"2. Como Prever Crises: {dados.get('prever', 'N√£o informado.')}")
                print(f"3. Como Lidar com Crises: {dados.get('lidar', 'N√£o informado.')}")
                print(f"4. Interesses (Motiva√ß√£o): {dados.get('interesses', 'N√£o informado.')}")
                print(f"5. Desinteresses (Distra√ß√£o): {dados.get('desinteresses', 'N√£o informado.')}")
                if input("\nDeseja atualizar estas informa√ß√µes? (s/n): ").lower() == 's':
                    pai['dados_filhos'][nome_filho] = {
                        'gatilhos': input("Gatilhos de crise: "), 'prever': input("Como prever crises: "),
                        'lidar': input("Como lidar com crises: "), 'interesses': input("Interesses: "),
                        'desinteresses': input("Desinteresses: ")
                    }
                    print("Informa√ß√µes atualizadas com sucesso!")
            else:
                print("Filho n√£o encontrado na sua lista.")
            aguardar_enter()

        elif opcao == '2':
            print("\n--- Informar/Atualizar Neurodiverg√™ncia ---")
            print(f"A informa√ß√£o de neurodiverg√™ncia registrada para seus filhos √©: \"{pai.get('neurodivergencias', 'Nenhuma')}\"")
            nova_info = input("Digite a nova informa√ß√£o (ou pressione Enter para manter a atual): ")
            if nova_info:
                pai['neurodivergencias'] = nova_info
                print("Informa√ß√£o de neurodiverg√™ncia atualizada com sucesso!")
            else:
                print("Nenhuma altera√ß√£o foi feita.")
            aguardar_enter()

        elif opcao == '3':
            print("\n--- Agenda Escolar ---")
            if not pai['filhos_nomes']:
                print("Nenhum filho cadastrado para visualizar a agenda.")
            else:
                aluno_encontrado = encontrar_aluno_por_nome(pai['filhos_nomes'][0])
                if aluno_encontrado:
                    agenda_turma = [item for item in agenda_global if item['turma'] == aluno_encontrado['turma']]
                    if not agenda_turma: print("Nenhum evento na agenda para a turma do seu filho.")
                    else:
                        for item in agenda_turma:
                            print(f"\n--- Evento na Agenda ---")
                            print(f"  Data: {item['data']} | Turma: {item['turma']}")
                            print(f"  Tipo: {item['tipo']}")
                            print(f"  Descri√ß√£o:")
                            print(formatar_texto_longo(item['descricao']))
                else: print("N√£o foi poss√≠vel encontrar os dados do aluno para exibir a agenda.")
            aguardar_enter()

        elif opcao == '4':
            nome_filho = input("Qual filho voc√™ quer ver o painel de sentimentos? ")
            aluno = encontrar_aluno_por_nome(nome_filho)
            if aluno and aluno['nome_completo'] in pai['filhos_nomes']:
                print(f"\n--- Painel de Sentimentos de {nome_filho} ---")
                if not aluno['painel_sentimentos']: print("Seu filho ainda n√£o registrou nenhum sentimento.")
                else:
                    for registro in aluno['painel_sentimentos']:
                        print(f"\n--- Registro de {registro['data']} ---")
                        print(f"  Sentimento: {registro['emoji']}")
                        print(f"  Relato:")
                        print(formatar_texto_longo(registro['texto'], recuo_inicial='    ', recuo_subsequente='    '))
            else: print("Aluno n√£o encontrado ou n√£o associado a voc√™.")
            aguardar_enter()

        elif opcao == '5':
            nome_filho = input("Qual filho voc√™ quer ver o desempenho? ")
            aluno = encontrar_aluno_por_nome(nome_filho)
            if aluno and aluno['nome_completo'] in pai['filhos_nomes']:
                exibir_grafico_desempenho(aluno)
            else: print("Aluno n√£o encontrado ou n√£o associado a voc√™.")

        elif opcao == '6':
            print("\n--- Perguntas dos Filhos ---")
            perguntas_filhos = [p for p in perguntas_alunos if p['aluno'] in pai['filhos_nomes']]
            if not perguntas_filhos: print("Nenhuma pergunta de seus filhos no momento.")
            else:
                for p in perguntas_filhos:
                    resposta = p.get('resposta', 'Aguardando resposta do professor...')
                    print(f"\n--- Pergunta de {p['aluno']} ---")
                    print(f"  Pergunta:")
                    print(formatar_texto_longo(p['pergunta']))
                    print(f"  Resposta:")
                    print(formatar_texto_longo(resposta))
            aguardar_enter()

        elif opcao == '7':
            print("\n--- F√≥rum: Pais e Professores ---")
            if not forum_geral: print("Nenhuma postagem no f√≥rum ainda.")
            for post in forum_geral:
                print(f"\n--------------------")
                print(f"  Autor: {post['autor']} ({post['tipo_autor']})")
                print(f"  Post:")
                print(formatar_texto_longo(post['post']))
                print(f"--------------------")
            aguardar_enter()

        elif opcao == '8':
            print("\n--- Nova Postagem no F√≥rum ---")
            post_texto = input("Sua mensagem: ")
            forum_geral.append({'autor': pai['nome_completo'], 'tipo_autor': 'Pai/M√£e', 'post': post_texto})
            print("Mensagem postada no f√≥rum!")
            aguardar_enter()

        elif opcao == '9':
            print("\n--- Caixa de Mensagens ---")
            mensagens_recebidas = False
            for msg in mensagens_pais_professores:
                if msg['destinatario'] == pai['nome_completo']:
                    print(f"\n--- Mensagem Recebida ---")
                    print(f"  De: {msg['remetente']}")
                    print(f"  Mensagem:")
                    print(formatar_texto_longo(msg['mensagem']))
                    mensagens_recebidas = True
            if not mensagens_recebidas: print("Nenhuma mensagem nova recebida.")
            aguardar_enter()

        elif opcao == '10':
            print("\n--- Enviar Nova Mensagem ---")
            dest = input("Nome do Professor destinat√°rio: ")
            if any(p['nome_completo'].lower() == dest.lower() for p in usuarios_professores):
                msg_texto = input("Sua mensagem: ")
                mensagens_pais_professores.append({'remetente': pai['nome_completo'], 'destinatario': dest, 'mensagem': msg_texto})
                print("Mensagem enviada!")
            else: print("Professor n√£o encontrado.")
            aguardar_enter()

        elif opcao == '0':
            break

def tela_professor(professor):
    while True:
        limpar_tela()
        print(f"========= PAINEL DOS PROFESSORES - Ol√°, Prof. {professor['nome_completo']} =========")
        print("\n--- Gest√£o Acad√™mica ---")
        print("1. Acessar dados de apoio de um aluno")
        print("2. Definir Pesos das Avalia√ß√µes")
        print("3. Lan√ßamento Sequencial de Notas")
        print("4. Ver Desempenho Detalhado de um aluno")
        print("5. Inserir item na Agenda")
        print("6. Ver Painel de Sentimentos de um aluno")
        print("\n--- Comunica√ß√£o ---")
        print("7. Visualizar F√≥rum")
        print("8. Postar no F√≥rum")
        print("9. Ler Mensagens")
        print("10. Enviar Mensagem para Pai/M√£e")
        print("11. Responder Perguntas de Alunos")
        print("\n0. Logout (Sair)")
        opcao = input("Escolha uma op√ß√£o: ")

        if opcao == '1':
            nome_aluno = input("Nome do aluno que deseja consultar: ")
            aluno = encontrar_aluno_por_nome(nome_aluno)
            if aluno:
                pai_aluno = encontrar_pai_por_filho(aluno['nome_completo'])
                if pai_aluno:
                    dados_apoio = pai_aluno['dados_filhos'].get(aluno['nome_completo'], {})
                    print(f"\n--- Dados de Apoio para {aluno['nome_completo']} (Fornecido por {pai_aluno['nome_completo']}) ---")
                    print(f"Neurodiverg√™ncias: {pai_aluno.get('neurodivergencias', 'N/A')}")
                    print(f"Gatilhos de Crises:")
                    print(formatar_texto_longo(dados_apoio.get('gatilhos')))
                    print(f"Como Prever Crises:")
                    print(formatar_texto_longo(dados_apoio.get('prever')))
                    print(f"Como Lidar com Crises:")
                    print(formatar_texto_longo(dados_apoio.get('lidar')))
                    print(f"Interesses (Motiva√ß√£o):")
                    print(formatar_texto_longo(dados_apoio.get('interesses')))
                    print(f"Desinteresses (Distra√ß√£o):")
                    print(formatar_texto_longo(dados_apoio.get('desinteresses')))
                else: print("Pais do aluno n√£o encontrados.")
            else: print("Aluno n√£o encontrado.")
            aguardar_enter()

        elif opcao == '2':
            print("\n--- Defini√ß√£o de Pesos para M√©dia Ponderada ---")
            turma_pesos = input(f"Para qual turma deseja definir os pesos? ({', '.join(professor['turmas'])}): ")
            if turma_pesos in professor['turmas']:
                try:
                    peso_provas = float(input("Digite o peso para as Provas (ex: 5): "))
                    peso_atividades = float(input("Digite o peso para as Atividades (ex: 3): "))
                    peso_participacao = float(input("Digite o peso para a Participa√ß√£o (ex: 2): "))
                    pesos_por_turma[turma_pesos] = {'provas': peso_provas, 'atividades': peso_atividades, 'participacao': peso_participacao}
                    print("Pesos definidos com sucesso para a turma", turma_pesos)
                except ValueError: print("Erro: Por favor, insira apenas n√∫meros.")
            else: print("Voc√™ n√£o leciona para esta turma.")
            aguardar_enter()

        elif opcao == '3':
            nome_aluno = input("Nome do aluno para o lan√ßamento de notas: ")
            aluno = encontrar_aluno_por_nome(nome_aluno)
            if aluno:
                print(f"\n--- Lan√ßamento Sequencial para {aluno['nome_completo']} ---")
                print("Dica: Pressione Enter sem digitar nada para pular um campo.")
                nota_str = input("-> Nova nota de Prova (0-10): ")
                if nota_str:
                    try:
                        aluno['desempenho']['provas'].append(float(nota_str))
                        print("   ...nota da prova registrada.")
                    except ValueError: print("   ...valor inv√°lido, pulando.")
                nota_str = input("-> Nova nota de Atividade (0-10): ")
                if nota_str:
                    try:
                        aluno['desempenho']['atividades'].append(float(nota_str))
                        print("   ...nota da atividade registrada.")
                    except ValueError: print("   ...valor inv√°lido, pulando.")
                nota_str = input("-> Nova nota de Participa√ß√£o (0-10): ")
                if nota_str:
                    try:
                        aluno['desempenho']['participacao'].append(float(nota_str))
                        print("   ...nota de participa√ß√£o registrada.")
                    except ValueError: print("   ...valor inv√°lido, pulando.")
                faltas_str = input("-> Adicionar Faltas (ex: 2): ")
                if faltas_str:
                    try:
                        aluno['desempenho']['faltas'] += int(faltas_str)
                        print("   ...faltas adicionadas.")
                    except ValueError: print("   ...valor inv√°lido, pulando.")
                print("\nLan√ßamento conclu√≠do para este aluno.")
            else: print("Aluno n√£o encontrado.")
            aguardar_enter()

        elif opcao == '4':
            nome_aluno = input("Qual aluno voc√™ quer ver o desempenho? ")
            aluno = encontrar_aluno_por_nome(nome_aluno)
            if aluno: exibir_grafico_desempenho(aluno)
            else: print("Aluno n√£o encontrado.")

        elif opcao == '5':
            print("\n--- Inserir na Agenda Global ---")
            turma_agenda = input(f"Para qual turma? ({', '.join(professor['turmas'])}): ")
            if turma_agenda in professor['turmas']:
                agenda_global.append({'turma': turma_agenda, 'data': input("Data (dd/mm/aaaa): "), 'tipo': input("Tipo (Prova/Evento): "), 'descricao': input("Descri√ß√£o: ")})
                print("Item adicionado √† agenda!")
            else: print("Voc√™ n√£o leciona para esta turma.")
            aguardar_enter()

        elif opcao == '6':
            nome_aluno = input("Qual aluno voc√™ quer ver o painel de sentimentos? ")
            aluno = encontrar_aluno_por_nome(nome_aluno)
            if aluno:
                print(f"\n--- Painel de Sentimentos de {nome_aluno} ---")
                if not aluno['painel_sentimentos']: print("Este aluno ainda n√£o registrou nenhum sentimento.")
                else:
                    for registro in aluno['painel_sentimentos']:
                        print(f"\n--- Registro de {registro['data']} ---")
                        print(f"  Sentimento: {registro['emoji']}")
                        print(f"  Relato:")
                        print(formatar_texto_longo(registro['texto'], recuo_inicial='    ', recuo_subsequente='    '))
            else: print("Aluno n√£o encontrado.")
            aguardar_enter()

        elif opcao == '7':
            print("\n--- F√≥rum: Pais e Professores ---")
            if not forum_geral: print("Nenhuma postagem no f√≥rum ainda.")
            else:
                for post in forum_geral:
                    print(f"\n--------------------\n  Autor: {post['autor']} ({post['tipo_autor']})\n  Post:")
                    print(formatar_texto_longo(post['post']))
                    print(f"--------------------")
            aguardar_enter()

        elif opcao == '8':
            print("\n--- Nova Postagem no F√≥rum ---")
            post_texto = input("Sua mensagem: ")
            forum_geral.append({'autor': professor['nome_completo'], 'tipo_autor': 'Professor', 'post': post_texto})
            print("Mensagem postada no f√≥rum!")
            aguardar_enter()

        elif opcao == '9':
            print("\n--- Caixa de Mensagens ---")
            mensagens_recebidas = False
            for msg in mensagens_pais_professores:
                if msg['destinatario'] == professor['nome_completo']:
                    print(f"\n--- Mensagem Recebida ---")
                    print(f"  De: {msg['remetente']}")
                    print(f"  Mensagem:")
                    print(formatar_texto_longo(msg['mensagem']))
                    mensagens_recebidas = True
            if not mensagens_recebidas: print("Nenhuma mensagem nova recebida.")
            aguardar_enter()

        elif opcao == '10':
            print("\n--- Enviar Nova Mensagem ---")
            dest = input("Nome do Pai/M√£e destinat√°rio: ")
            if any(p['nome_completo'].lower() == dest.lower() for p in usuarios_pais):
                msg_texto = input("Sua mensagem: ")
                mensagens_pais_professores.append({'remetente': professor['nome_completo'], 'destinatario': dest, 'mensagem': msg_texto})
                print("Mensagem enviada!")
            else: print("Pai/M√£e n√£o encontrado.")
            aguardar_enter()

        elif opcao == '11':
            print("\n--- Perguntas dos Alunos ---")
            if not perguntas_alunos: print("Nenhuma pergunta no momento.")
            else:
                perguntas_pendentes = [p for p in perguntas_alunos if 'resposta' not in p]
                if not perguntas_pendentes: print("Todas as perguntas j√° foram respondidas.")
                else:
                    print("--- Perguntas Pendentes ---")
                    for i, p in enumerate(perguntas_pendentes):
                        print(f"\n{i+1}. Pergunta de {p['aluno']}:")
                        print(formatar_texto_longo(p['pergunta'], recuo_inicial='   '))
                    try:
                        idx_pendente = int(input("\nDigite o n√∫mero da pergunta a responder (0 para voltar): ")) - 1
                        if 0 <= idx_pendente < len(perguntas_pendentes):
                            pergunta_original = perguntas_pendentes[idx_pendente]
                            resposta = input("Sua resposta: ")
                            pergunta_original['resposta'] = resposta
                            print("Resposta enviada!")
                        elif idx_pendente != -1: print("N√∫mero inv√°lido.")
                    except (ValueError, IndexError): print("Entrada inv√°lida.")
            aguardar_enter()

        elif opcao == '0':
            break

def tela_aluno(aluno):
    lista_desafios = [
        "Respira√ß√£o do Bal√£o: Finja que sua barriga √© um bal√£o...", "Ca√ßador de Cores: Encontre 5 coisas azuis...", "Desenho Cego: Desenhe como voc√™ se sente de olhos fechados.",
        "O Som Misterioso: Fique em sil√™ncio por um minuto e ou√ßa...", "Poema de 3 Linhas: Crie um poema sobre seu animal favorito.", "Caminhada Atenta: D√™ 10 passos prestando aten√ß√£o total.",
        "Super-Her√≥i da Gentileza: Pense em algo legal para fazer por algu√©m.", "Escultor de Nuvens: Encontre formas nas nuvens.", "Mapa do Tesouro Criativo: Desenhe um mapa de um lugar m√°gico.",
        "Cinco Coisas Legais: Escreva 5 coisas que voc√™ acha legal em voc√™ mesmo."
    ]
    while True:
        limpar_tela()
        print(f"========= PAINEL DO ALUNO - Ol√°, {aluno['nome_completo']} e {aluno['nome_avatar']}! =========")
        print(f"Seu amigo avatar: {aluno['imagem_avatar']}")

        progresso_sol = aluno.get('desafios_sol', 0)
        print(f"\nSeu Sol Brilha: {'‚òÄÔ∏è' * progresso_sol}{'‚ö´' * (10 - progresso_sol)} [{progresso_sol}/10 desafios]")

        print("\n--- Menu de Op√ß√µes ---")
        print("1. Como estou me sentindo hoje?")
        print("2. Desafios do Sol")
        print("3. Fazer uma pergunta para o professor")
        print("4. Ver minhas Perguntas e Respostas")
        print("5. Ver meu Desempenho")
        print("6. Galeria de Cria√ß√µes")
        print("0. Logout (Sair)")
        opcao = input("Escolha uma op√ß√£o: ")

        if opcao == '1':
            limpar_tela()
            print("\n--- Como voc√™ se sente hoje? ---")
            opcoes_sentimentos = {
                '1': ('üòä', 'Feliz / Contente'), '2': ('üò¢', 'Triste / Chateado'), '3': ('üò†', 'Bravo / Irritado'),
                '4': ('üò¥', 'Cansado / Com Sono'), '5': ('ü§î', 'Pensativo / Confuso'), '6': ('üéâ', 'Animado / Empolgado'),
                '7': ('üòü', 'Preocupado(a)')
            }
            for num, (emoji, desc) in opcoes_sentimentos.items(): print(f"{num}. {emoji} {desc}")
            escolha = input("\nEscolha um n√∫mero que representa seu sentimento (ou 0 para voltar): ")
            if escolha in opcoes_sentimentos:
                emoji_escolhido, _ = opcoes_sentimentos[escolha]
                texto = input(f"\nVoc√™ escolheu {emoji_escolhido}. Conte um pouco mais sobre o porqu√™:\n")
                data_hoje = time.strftime("%d/%m/%Y")
                aluno['painel_sentimentos'].append({'data': data_hoje, 'emoji': emoji_escolhido, 'texto': texto})
                print("\nObrigado por compartilhar seus sentimentos!")
            elif escolha != '0': print("Op√ß√£o inv√°lida.")
            aguardar_enter()

        elif opcao == '2':
            limpar_tela()
            print("--- Desafios do Sol para Acalmar e Criar ---")
            if progresso_sol >= 10: print("\nParab√©ns! Voc√™ completou todos os desafios! ‚òÄÔ∏è‚òÄÔ∏è‚òÄÔ∏è‚òÄÔ∏è‚òÄÔ∏è‚òÄÔ∏è‚òÄÔ∏è‚òÄÔ∏è‚òÄÔ∏è‚òÄÔ∏è")
            else:
                for i, desafio in enumerate(lista_desafios):
                    print(f"\n{i + 1}. Desafio:")
                    print(formatar_texto_longo(desafio))
                try:
                    escolha = input("\nEscolha o n√∫mero de um desafio para completar (ou 0 para voltar): ")
                    if escolha != '0' and 1 <= int(escolha) <= 10:
                        aluno['desafios_sol'] += 1
                        print("\nDesafio conclu√≠do! Seu sol brilhou mais forte!")
                    elif escolha != '0': print("N√∫mero de desafio inv√°lido.")
                except ValueError: print("Por favor, digite um n√∫mero.")
            aguardar_enter()

        elif opcao == '3':
            print("\n--- Fazer uma Pergunta ---")
            pergunta_texto = input("Qual √© a sua d√∫vida? (pressione Enter para enviar):\n")
            perguntas_alunos.append({'aluno': aluno['nome_completo'], 'pergunta': pergunta_texto})
            print("Pergunta enviada ao professor!")
            aguardar_enter()

        elif opcao == '4':
            limpar_tela()
            print("--- Minhas Perguntas e Respostas ---")
            minhas_perguntas = [p for p in perguntas_alunos if p['aluno'] == aluno['nome_completo']]
            if not minhas_perguntas:
                print("\nVoc√™ ainda n√£o enviou nenhuma pergunta.")
            else:
                for p in minhas_perguntas:
                    resposta = p.get('resposta', 'Aguardando resposta do professor...')
                    print(f"\n--------------------")
                    print("Sua Pergunta:")
                    print(formatar_texto_longo(p['pergunta']))
                    print("\nResposta do Professor:")
                    print(formatar_texto_longo(resposta))
                    print(f"--------------------")
            aguardar_enter()

        elif opcao == '5':
            exibir_grafico_desempenho(aluno)

        elif opcao == '6':
            print("\n--- Galeria de Cria√ß√µes ---\nEm desenvolvimento! Em breve voc√™ poder√° compartilhar suas cria√ß√µes aqui.")
            aguardar_enter()

        elif opcao == '0':
            break

# ==============================================================================
# FUN√á√ÉO PRINCIPAL (MAIN LOOP)
# ==============================================================================

def main():
    if not usuarios_professores:
      usuarios_professores.append({'tipo': 'professor', 'nome_completo': 'Prof. Alan Turing', 'instituicao': 'SENAI FATESG', 'turmas': ['IA-1A', 'IA-1B'], 'username': 'alan', 'senha': '123'})
    if not usuarios_pais:
      usuarios_pais.append({'tipo': 'pai', 'nome_completo': 'Maria Santos', 'filhos_nomes': ['Jo√£o Santos'], 'neurodivergencias': 'TDAH', 'instituicao': 'SENAI FATESG', 'username': 'maria', 'senha': '123', 'dados_filhos': {}})
    if not usuarios_alunos:
      usuarios_alunos.append({'tipo': 'aluno', 'nome_completo': 'Jo√£o Santos', 'nome_pais': 'Maria Santos', 'nome_avatar': 'Joca Coder', 'imagem_avatar': 'ü§ñ Rob√¥ Curioso', 'username': 'joao', 'senha': '123', 'painel_sentimentos': [], 'desempenho': {'provas': [7.5, 8.0], 'atividades': [10, 9], 'participacao': [8.0], 'faltas': 2}, 'desafios_sol': 0, 'turma': 'IA-1A'})
    if not pesos_por_turma:
        pesos_por_turma['IA-1A'] = {'provas': 5, 'atividades': 3, 'participacao': 2}

    while True:
        limpar_tela()
        print("#####################################")
        print("###     BEM-VINDO AO COGNIKIDS    ###")
        print("#####################################")
        print("\nPara come√ßar, quem √© voc√™?")
        print("1. Sou Pai/M√£e\n2. Sou Aluno(a)\n3. Sou Professor(a)\n\n0. Sair do Programa")
        tipo_acesso = input("Escolha o seu perfil: ")

        if tipo_acesso == '0':
            print("Obrigado por usar o CogniKids. At√© a pr√≥xima!")
            break

        if tipo_acesso not in ['1', '2', '3']:
            print("Op√ß√£o inv√°lida."); aguardar_enter(); continue

        primeiro_acesso = input("√â seu primeiro acesso? (s/n): ").lower()
        if primeiro_acesso == 's':
            if tipo_acesso == '1': cadastrar_pai()
            elif tipo_acesso == '2': cadastrar_aluno()
            elif tipo_acesso == '3': cadastrar_professor()
        elif primeiro_acesso == 'n':
            usuario_logado = realizar_login()
            if usuario_logado:
                tipo_correto = ((tipo_acesso == '1' and usuario_logado['tipo'] == 'pai') or (tipo_acesso == '2' and usuario_logado['tipo'] == 'aluno') or (tipo_acesso == '3' and usuario_logado['tipo'] == 'professor'))
                if tipo_correto:
                    if usuario_logado['tipo'] == 'pai': tela_pais(usuario_logado)
                    elif usuario_logado['tipo'] == 'professor': tela_professor(usuario_logado)
                    elif usuario_logado['tipo'] == 'aluno': tela_aluno(usuario_logado)
                else: print("\nErro: Voc√™ tentou fazer login com um tipo de perfil diferente do seu cadastro."); aguardar_enter()
        else: print("Op√ß√£o inv√°lida. Use 's' para sim ou 'n' para n√£o."); aguardar_enter()

# --- Ponto de Entrada do Programa ---
if __name__ == "__main__":
    main()
```
